```python
import matplotlib.pyplot as plt

cantidadF = df['VACUNA'][df['GENERO'] == 'F'].count()
cantidadM = df['VACUNA'][df['GENERO'] == 'M'].count()
cantidadN = df['VACUNA'][df['GENERO'] == 'N'].count()

cantidades = [cantidadF, cantidadM, cantidadN]
genero = ['F', 'M', 'N']
colores = ['pink','cyan','black']

plt.pie(cantidades, labels = genero, colors = colores, startangle = 45, autopct = '%1.2f%%')
plt.title('Porcentajes por género')
plt.legend(loc= 'best')
plt.axis('equal')
plt.show()
```

* La diferencia entre el porcentaje entre hombres y mujeres es de 3.62%.
